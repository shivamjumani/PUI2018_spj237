```python
import numpy as np
import pylab as pl
%pylab inline
```

```python
np.random.seed(999)
#fixes the values
```

```python
pl.hist(np.random.randn(100))
```

```python
histos = np.zeros((5000, 100))
#creating the empty array
for i in range(0, 5000):
    histos[i] = np.random.randn(100)
print(histos.shape)

pl.hist(histos[1])
#print one of the histograms

#do this 5000 times. each time give a sample of 100.
#need to create an array(container), one of each histogram sample, and then put the histograms in there
```

```python
def ztest(sample, mu=0, std=1):
    xaverage = sample.mean()
    N = sample.size
    return (mu - xaverage) / (std / np.sqrt(N))

#making it default when you set mu=0, std=1, unless it is specified.
```

```python
ztest(histos[2])
#histos[1] is the 'sample' that you have taken, which is defined in the function
```

```python
zlist = []
for i,s in enumerate(histos):
    zlist.append(ztest(s))
    
#zlist is a list, not an array. It is more convenient in this case. that's why the append. makes sense making a list if you don't know the size of hte list coz you're just throwing things in there. 
#use array if you know the size. we do, but we still used the list
```

```python
pl.hist(zlist)
```

```python
np.mean(zlist)
```

```python
np.std(zlist)
```

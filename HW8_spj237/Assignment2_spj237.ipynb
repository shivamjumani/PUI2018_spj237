IDEA

More tourists ride Citibike on the weekends compared to that on weekdays. I have assumed that customers are tourists and subscribers are local.

NULL HYPOTHESIS

The fraction of customers biking on weekends is the same or lower than the fraction of those biking on weekdays.

ALTERNATIVE HYPOTHESIS

The fraction of customers biking on weekends is the greater than the fraction of those biking on weekdays.

$H_0: \frac{C_{\mathrm{weekend}}}{C_{\mathrm{total}}} <= \frac{C_{\mathrm{weekday}}}{C_{\mathrm{total}}}$

$H_1: \frac{C_{\mathrm{weekend}}}{C_{\mathrm{total}}} > \frac{C_{\mathrm{weekday}}}{C_{\mathrm{total}}}$

I will use a significance level  $\alpha=0.05$

# Chi square test

```python
def evalChisq(values):
    '''Evaluates the chi sq from a contingency value
    Arguments:
    values: 2x2 array or list, the contingengy table
    '''
    if not (len(values.shape) == 2 and values.shape == (2,2)):
        print ("must pass a 2D array")
        return -1
    values = np.array(values)
    E = np.empty_like(values)
    for j in range(len(values[0])):
        for i in range(2):
            
            E[i][j] = ((values[i,:].sum() * values[:,j].sum()) / 
                        (values).sum())
    return ((values - E)**2 / E).sum()
```
